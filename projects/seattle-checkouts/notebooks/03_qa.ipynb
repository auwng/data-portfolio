# Title Cleaning

### Assess discrepancies between how book/ ebook titles, authors vary. try to get them as close as possible for string matching between the two

```python
#import data

book_checkouts = physical_checkouts[
    (physical_checkouts['materialtype'] == 'BOOK')]

ebook_checkouts = digital_checkouts[
    (digital_checkouts['materialtype'] == 'EBOOK')]
```

### check book title formats

```python
#identify common separators and how often they're used for formatting book title metadata

import re
from collections import Counter

def inspect_separator(sep, n=20):
    escaped = re.escape(sep)
    total_distinct = book_checkouts['title'].nunique()
    matches = book_checkouts[book_checkouts['title'].str.contains(escaped, na=False, regex=False)]
    distinct_matches = matches['title'].nunique()
    pct = distinct_matches / total_distinct * 100
    
    print(f"=== Titles containing '{sep}' ===")
    print(f"Distinct titles: {distinct_matches:,} of {total_distinct:,} ({pct:.1f}%)\n")
    
    sample = matches['title'].drop_duplicates().sample(min(n, distinct_matches))
    for title in sample:
        parts = [p.strip() for p in title.split(sep)]
        for i, part in enumerate(parts):
            print(f"  Part {i+1}: {part[:80]}")
        print()

for sep in ['/', ':', ';', ',', '||']:
    inspect_separator(sep)
    print('=' * 60)
```

```python
# evaluate how often splitting title on / will be correct

# Split on / and evaluate each part
def evaluate_slash_splits(df, title_col='title'):
    titles_with_slash = df[df[title_col].str.contains('/', na=False, regex=False)][title_col].drop_duplicates()
    
    total = len(titles_with_slash)
    print(f"Total distinct titles with '/': {total:,}\n")
    
    # Heuristics for when / is likely IN the title vs a separator
    likely_in_title = titles_with_slash[
        titles_with_slash.str.contains(r'\d+/\d+', na=False)  # fractions like 1/2
        | titles_with_slash.str.contains(r'[A-Z]/[A-Z]', na=False)  # acronyms like AC/DC
        | titles_with_slash.str.contains(r'\w/\w', na=False)  # no spaces around slash
    ]
    
    likely_separator = titles_with_slash[
        titles_with_slash.str.contains(r'\s/\s', na=False)  # spaces on both sides
    ]

    print(f"Likely separator (spaces on both sides):  {len(likely_separator):,} ({len(likely_separator)/total*100:.1f}%)")
    print(f"Likely in title (no spaces / acronym):    {len(likely_in_title):,} ({len(likely_in_title)/total*100:.1f}%)")
    print(f"Ambiguous (other):                        {total - len(likely_separator) - len(likely_in_title):,} ({(total - len(likely_separator) - len(likely_in_title))/total*100:.1f}%)")

    print(f"\n=== Sample: likely separator ===")
    for title in likely_separator.sample(min(10, len(likely_separator))):
        parts = title.split('/')
        print(f"  Part 1: {parts[0].strip()[:60]}")
        print(f"  Part 2: {'/'.join(parts[1:]).strip()[:60]}")
        print()

    print(f"=== Sample: likely in title ===")
    for title in likely_in_title.sample(min(10, len(likely_in_title))):
        print(f"  {title[:80]}")
        print()

    #return ambiguous responses

    ambiguous = titles_with_slash[
        ~titles_with_slash.isin(likely_separator) & 
        ~titles_with_slash.isin(likely_in_title)
    ]
    
    print(f"=== Sample: ambiguous ===")
    for title in ambiguous.sample(min(10, len(ambiguous))):
        print(f"  {title[:80]}")
        print()
    
    return ambiguous

ambiguous = evaluate_slash_splits(book_checkouts)
```

### only returning the first part of the book title before ' / ' separator 
### addresses 99.8% of its use as the most common separator character.

## EBook Title Cleaning - check frequency of parentheses containing noise

```python
processed_dir = '/Users/audriswong/data-portfolio/projects/seattle-checkouts/data/processed'

#import data
book_checkouts = pd.read_parquet(f'{processed_dir}/book_checkouts.parquet').rename(columns={'title_clean': 'title'}) 
ebook_checkouts = pd.read_parquet(f'{processed_dir}/ebook_checkouts.parquet').rename(columns={'title_clean': 'title'})
```

```python
##worth removing extra content in the parentheses?  e.g. (Pulitzer Prize, Oprah's Book Club)

import re
from collections import Counter

# Extract all parenthetical content from ebook titles
parens_content = []

for title in ebook_checkouts['title'].dropna():
    matches = re.findall(r'\(([^)]+)\)', str(title))
    parens_content.extend(matches)

parens_counts = Counter(parens_content)
print(f"=== Ebook Title Breakdown ===")
print(f"Total parenthetical instances: {len(parens_content):,}")
print(f"Distinct parenthetical values: {len(parens_counts):,}")
print("\n=== Most common parenthetical content ===")
for content, count in parens_counts.most_common(30):
    print(f"{count:>8,} | {content[:80]}")

#book titles
parens_content = []

for title in book_checkouts['title'].dropna():
    matches = re.findall(r'\(([^)]+)\)', str(title))
    parens_content.extend(matches)

parens_counts = Counter(parens_content)
print(f"=== Book Title Breakdown ===")
print(f"Total parenthetical instances: {len(parens_content):,}")
print(f"Distinct parenthetical values: {len(parens_counts):,}")
print("\n=== Most common parenthetical content ===")
for content, count in parens_counts.most_common(30):
    print(f"{count:>8,} | {content[:80]}")
```

```python
# Classify parenthetical content into categories
def classify_paren(content):
    c = content.lower().strip()
    
    if re.search(r'\d{4}', c):
        return 'Year/Date'
    elif any(w in c for w in ['prize', 'award', 'winner', 'pulitzer', 'booker', 'newbery', 'caldecott', 'national book']):
        return 'Award/Prize'
    elif any(w in c for w in ['book club', 'reese', 'oprah', 'good morning america', 'gma', 'today show']):
        return 'Book Club'
    elif any(w in c for w in ['audio', 'unabridged', 'abridged', 'narrated', 'read by']):
        return 'Audio Details'
    elif any(w in c for w in ['bonus', 'excerpt', 'preview', 'includes', 'with ', 'enhanced', 'extra']):
        return 'Bonus Content'
    elif any(w in c for w in ['book 1', 'book 2', 'book 3', 'vol', 'volume', 'part ', 'series', '#']):
        return 'Series/Volume'
    elif any(w in c for w in ['edition', 'revised', 'updated', 'expanded', 'anniversary']):
        return 'Edition Info'
    elif any(w in c for w in ['translation', 'translated', 'adapted']):
        return 'Translation'
    elif any(w in c for w in ['illustrated', 'illustrations', 'pictures']):
        return 'Illustration Info'
    elif any(w in c for w in ['novel', 'memoir', 'story', 'tales', 'poems', 'essays']):
        return 'Likely Title Content'
    elif len(c.split()) <= 3:
        return 'Short — likely title content'
    else:
        return 'Other/Unclear'

# Apply classification
paren_classified = Counter()
paren_examples   = {}

for title in ebook_checkouts['title'].dropna():
    matches = re.findall(r'\(([^)]+)\)', str(title))
    for match in matches:
        category = classify_paren(match)
        paren_classified[category] += 1
        if category not in paren_examples:
            paren_examples[category] = []
        if len(paren_examples[category]) < 5:
            paren_examples[category].append(f"{match[:60]} <- {title[:60]}")

total = sum(paren_classified.values())
print(f"=== Parenthetical Content Categories ===\n")
for category, count in sorted(paren_classified.items(), key=lambda x: x[1], reverse=True):
    pct = count / total * 100
    print(f"{count:>8,} ({pct:>5.1f}%)  {category}")
    for ex in paren_examples.get(category, []):
        print(f"            e.g. {ex}")
    print()
```

### not worth removing all content within ().  ~80% of parentheses content within ebook titles are actually parts of the title
